```python
import pandas as pd
import csv
from csv import DictReader
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/Bungeetech/internship-test2/master/input/question-1/main.csv")
size = len(df2)
df2.head(10)
```

```python
df2.iloc[0][1]
a = size%10

print(a)
```

```python
f =open("answer-1.csv","w", newline="") 
fild = ['Year','Population',		'Violent',	'Property',	'Murder',	'Forcible_Rape'	,'Robbery','Aggravated_assault',	'Burglary','Larceny_Theft','Vehicle_Theft']
writer = csv.DictWriter(f,fieldnames=fild)
writer.writeheader()

row_counter  =0

for i in range(0,6):
  values = [0]*20
  for j in range(0,10) :
    if i*10+j<size:
      row_counter = row_counter+1
      for k in range(0,10):
        #print(i,'  ',j,"  ",k)
        values[k]= values[k]+df2.iloc[i*10+j][k+2]
        
      p =df2.iloc[i*10+j-1][1]
  writer.writerow({'Year':df2.iloc[i*10][0],'Population':p,		'Violent':values[1],	'Property':values[2],	'Murder':values[3],	'Forcible_Rape':values[4]	,'Robbery':values[5],'Aggravated_assault':values[6],	'Burglary':values[7],'Larceny_Theft':values[8],'Vehicle_Theft':values[9]})
  


f.close()    
```
