```python
import json
import pandas as pd
import numpy as np
import csv
from csv import DictReader
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/Bungeetech/internship-test2/master/input/question-2/main.csv")
size = len(df2)
df2.head(10)
```

```python
f =open("answer-2.csv","w", newline="") 
fild = ['occupation',"min","max"]
writer = csv.DictWriter(f,fieldnames=fild)
writer.writeheader()

row_counter  =0
d = {}
for i in range(0,943):
  k  = str(df2.iloc[i][3])
  
  if k in d.keys():
    miin = int(d[k][0])
    if miin>df2.iloc[i][1]:
      d[k][0] = df2.iloc[i][1]
    if d[k][1]<int(df2.iloc[i][1]):  
      d[k][1] = df2.iloc[i][1]
  else :
    d[k]=[df2.iloc[i][1],df2.iloc[i][1]]


for k in d.keys():
  writer.writerow({'occupation':k,"min":d[k][0],"max":d[k][1]})
  


f.close()    
```
